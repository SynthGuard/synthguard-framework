### Data loading

```python
df = bn.import_example(data='titanic')
```

### Data preprocessing

```python
print(df.head())
print(df.shape)

edges=[]
continuous_columns = []
#df_discrete = bn.discretize(df, edges, continuous_columns, max_iterations=1)


dfhot, dfnum = bn.df2onehot(df)
print(dfhot.head())
print(dfnum.head())
df = dfnum
```

### Structure learning 

```python
methods = [
    {'methodtype': 'hc', 'scoretype': 'bic'},
    {'methodtype': 'hc', 'scoretype': 'k2'},
    {'methodtype': 'hc', 'scoretype': 'bdeu'},
    {'methodtype': 'hc', 'scoretype': 'bds'},
    {'methodtype': 'hc', 'scoretype': 'aic'},
    {'methodtype': 'cs', 'scoretype': 'bic'}
]

# Fit the structure learning models
models = []
method_details = []
for method in methods:
    try:
        model = bn.structure_learning.fit(df, **method)
        models.append(model)
        method_details.append(method)  # Keep track of the method parameters
    except Exception as e:
        print(f"Error with method {method}: {e}")
        continue

# Compare models and determine the most accurate one
best_score = float('inf')
best_model = None
best_method = None

def calculate_accuracy(score_matrix):
    """Calculate accuracy from the confusion matrix."""
    DD = score_matrix[0, 0]
    DC = score_matrix[0, 1]
    CD = score_matrix[1, 0]
    CC = score_matrix[1, 1]
    total = DD + DC + CD + CC
    accuracy = (DD + CC) / total
    return accuracy

for i in range(len(models)):
    for j in range(i + 1, len(models)):
        try:
            scores, adjmat_diff = bn.compare_networks(models[i], models[j], verbose=0)
            print(f"Comparison between model {i} and model {j}: Score = {scores}")
            accuracy = calculate_accuracy(scores)
            if accuracy < best_score:
                best_score = accuracy
                best_model = models[i] if accuracy < best_score else models[j]
                best_method = method_details[i] if accuracy < best_score else method_details[j]
        except Exception as e:
            print(f"Error comparing models {i} and {j}: {e}")
            continue

print("Best model determined based on comparison scores:")
print(best_model)
print("Best model's method details:")
print(best_method)

# Optionally, visualize the best model
bn.plot(best_model)
```

### Paramter learning

```python
updated_model = bn.parameter_learning.fit(best_model, df)

new_df = bn.sampling(updated_model, n=10000)

updated_model2 = bn.parameter_learning.fit(updated_model, new_df)
print(bn.print_CPD(updated_model2))
```
